# 1. Introduction:

## 1.1  What is Exploratory Data Analysis?

-   Initial data exploration and analysis are the focus of the analytics discipline known as exploratory data analysis. We clean our data through exploratory data analysis and then use statistical and graphical techniques to better understand it. This way we can identify key columns in our machine-learning model, find patterns and insights in our data and understand data properties.

## 1.2  Football Analytics:

-   Over the past two decades, the influence of data analytics has been growing in every aspect of our lives: in businesses of every kind, but also in healthcare, media, and sports. Until a few years ago, football was thought to be immune from this trend. Now, the early adopters in the major football leagues are thriving thanks to the competitive advantage that investments in data analytics are beginning to provide them: Liverpool, AZ Alkmaar, and Brentford are just a few in the fast-growing list of successful case studies. In our view, the clubs that aren’t planning to jump on the analytics bandwagon run the risk of being left behind[1].<br>

-   The majority of publicly accessible football (soccer) statistics are only comprised of compiled information like Goals, Shots, Fouls, and Cards. This straightforward aggregation, which lacks context, might be misleading when evaluating performance or developing prediction models. For instance, a team with 10 long-range shots on target has a lesser chance of scoring than one with the same number of attempts from inside the box. However, the two teams will be evaluated similarly using criteria generated from this straightforward count of shots.<br>




-  Football uses data in three crucial areas: performance analysis, hiring, and strategy. Data is utilized to enhance pre- and post-game analysis of important player traits, set-piece patterns, creating chances, and team structure during different phases of play in performance analysis. This information can be used to assess team and player performance after a game in comparison to pre-set KPIs that were a part of the game's strategy.<br>
   In terms of recruitment, data is being used more and more to compare and evaluate players from other markets to sign new players and scout future transfer targets. This is utilized along with more conventional techniques and video. Traditionally, recruiting has been focused solely on enhancing the squad, although consideration of long-term financial advantage is growing.<br>

```python
# importing all necessary packges 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']=8,4
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Lodaing the Dataset


df17=pd.read_excel(r"EPL\17\Regular season\overall.xlsx")
df18=pd.read_excel(r"EPL\18\Regular season\overall.xlsx")
df19=pd.read_excel(r"EPL\19\Regular season\overall.xlsx")
df20=pd.read_excel(r"EPL\20\Regular season\overall.xlsx")
df21=pd.read_excel(r"EPL\21\Regular season\overall.xlsx")

af17=pd.read_excel(r"EPL\17\Regular season\homeaway17.xlsx")
af18=pd.read_excel(r"EPL\18\Regular season\homeaway18.xlsx")
af19=pd.read_excel(r"EPL\19\Regular season\homeaway19.xlsx")
af20=pd.read_excel(r"EPL\20\Regular season\homeaway20.xlsx")
af21=pd.read_excel(r"EPL\21\Regular season\homeaway21.xlsx")

pt17=pd.read_excel(r"EPL\17\Squad Pass type\passtype17.xlsx")
pt18=pd.read_excel(r"EPL\18\Squad Pass type\passtype18.xlsx")
pt19=pd.read_excel(r"EPL\19\Squad Pass type\passtype19.xlsx")
pt20=pd.read_excel(r"EPL\20\Squad Pass type\passtype20.xlsx")
pt21=pd.read_excel(r"EPL\21\Squad Pass type\passtype21.xlsx")


p17=pd.read_excel(r"EPL\17\Squad Passing\passing17.xlsx")
p18=pd.read_excel(r"EPL\18\Squad Passing\passing18.xlsx")
p19=pd.read_excel(r"EPL\19\Squad Passing\passing19.xlsx")
p20=pd.read_excel(r"EPL\20\Squad Passing\passing20.xlsx")
p21=pd.read_excel(r"EPL\21\Squad Passing\passing21.xlsx")
```

# Adding Columns to our Dataset


```python
# Adding "Season" column to the data for linking the two tables

df17['Season']=str(2017)
df18['Season']=str(2018)
df19['Season']=str(2019)
df20['Season']=str(2020)
df21['Season']=str(2021)

af17['Season']=str(2017)
af18['Season']=str(2018)
af19['Season']=str(2019)
af20['Season']=str(2020)
af21['Season']=str(2021)

pt17['Season']=str(2017)
pt18['Season']=str(2018)
pt19['Season']=str(2019)
pt20['Season']=str(2020)
pt21['Season']=str(2021)

p17['Season']=str(2017)
p18['Season']=str(2018)
p19['Season']=str(2019)
p20['Season']=str(2020)
p21['Season']=str(2021)
```

# Concatenating, Merging our dataset into one table


```python
# Concatenating our dataframes into one tables with the function `pd.concat

Overall_season = [df17,df18,df19,df20,df21]
overall_stats = pd.concat(Overall_season)

homeaway_overall_statsseason = [af17,af18,af19,af20,af21]
homeaway_stats = pd.concat(homeaway_overall_statsseason)

Passingtype_season=[pt17,pt18,pt19,pt20,pt21]
Passingtype_stats=pd.concat(Passingtype_season)

Passing_season=[p17,p18,p19,p20,p21]
passing_stats=pd.concat(Passing_season)
```

```python
Season = list(overall_liverpool['Season'])
xg_liv  = list(overall_liverpool['xG'])
gf_liv = list(overall_liverpool['GF'])
x = np.arange(len(Season))  # the label locations
width = 0.35  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, xg_liv, width, label='Expected(xG)')
rects2 = ax.bar(x + width/2, gf_liv, width, label='Goal For(GF)',color="red")
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Goals')
ax.set_title('Goals and Expected goals for liverpool')
ax.set_xticks(x,Season)
ax.legend()
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
fig.tight_layout()
plt.show()

Crystal_Palace=overall_stats.loc[overall_stats['Squad']=='Crystal Palace']
sea=list(Crystal_Palace['Season'])
xg_Cry = list(Crystal_Palace['xG'])
gf_Cry = list(Crystal_Palace['GF'])
x = np.arange(len(Season))  # the label locations
width = 0.35  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, xg_Cry, width, label='Expected(xG)')
rects2 = ax.bar(x + width/2, gf_Cry, width, label='Goal For(GF)',color="red")
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Goals')
ax.set_title('Goals and Expected goals for Crystal Palace')
ax.set_xticks(x,sea)
ax.legend()
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
fig.tight_layout()

plt.show()
```
